# GAA Score Collection

This notebook contains a GAA Pitch graphic that allows for capturing all shots that take place in a given Gaelic Football match. 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tkinter as tk
from tkinter import simpledialog
import matplotlib as mpl
```

```python
%matplotlib notebook
```

### Create Pitch

```python
coords = []
current_coords = []

ROOT = tk.Tk()


ROOT.withdraw()

def gaa_pitch(pitch, line, orientation,view):
    
  
    
    orientation = orientation
    view = view
    line = line
    pitch = pitch
    
    if orientation.lower().startswith("h"):
        
        if view.lower().startswith("h"):
            fig,ax = plt.subplots(figsize=(8.9,14.6))
            plt.xlim(89,146)
            plt.ylim(-1,89)
        else:
            fig,ax = plt.subplots(figsize=(14.5,8.8))
            plt.xlim(-1,146)
            plt.ylim(-1,89)
        ax.axis('off') # this hides the x and y ticks
    
        # side and goal lines #
        ly1 = [0,0,88,88,0]
        lx1 = [0,145,145,0,0]

        plt.plot(lx1,ly1,color=line,zorder=5)


            #goals#
        ly4 = [40.75,40.75,47.25,47.25]
        lx4 = [145,145.2,145.2,145]
        plt.plot(lx4,ly4,color=line,zorder=5)

        ly5 = [40.75,40.75,47.25,47.25]
        lx5 = [0,-0.2,-0.2,0]
        plt.plot(lx5,ly5,color=line,zorder=5)


           #6 yard boxes#
        ly6 = [37,37,51,51]
        lx6 = [145,140.5,140.5,145]
        plt.plot(lx6,ly6,color=line,zorder=5)

        ly7 = [37,37,51,51]
        lx7 = [0,4.5,4.5,0]
        plt.plot(lx7,ly7,color=line,zorder=5)
        
        #large rectangle
        ly88 = [34.5,34.5,53.5,53.5]
        lx88 = [145,131,131,145]
        plt.plot(lx88,ly88,color=line,zorder=5)

        ly87 = [34.5,34.5,53.5,53.5]
        lx87 = [0,14,14,0]
        plt.plot(lx87,ly87,color=line,zorder=5)

        
        #halfway small line
        ly97 = [39,49] 
        lx97 = [72.5,72.5]
        plt.plot(lx97,ly97,color=line,zorder=5)
        
        #peno lines
        ly90 = [43.5,44.5] 
        lx90 = [11,11]
        plt.plot(lx90,ly90,color=line,zorder=5)
        
        ly89 = [43.5,44.5] 
        lx89 = [136,136]
        plt.plot(lx89,ly89,color=line,zorder=5)

        circle1 = plt.Circle((124,44), 13,ls='solid',lw=1.5,color=line, fill=False, zorder=1,alpha=1)
        circle2 = plt.Circle((21,44), 13,ls='solid',lw=1.5,color=line, fill=False, zorder=1,alpha=1)
        circle3 = plt.Circle((52, 34), 9.15,ls='solid',lw=1.5,color=line, fill=False, zorder=2,alpha=1)

        
        #14 yard line across the pitch
        ly94 = [0,88] 
        lx94 = [14,14]
        plt.plot(lx94,ly94,color=line,zorder=5)
        
        ly99 = [0,88] 
        lx99 = [131,131]
        plt.plot(lx99,ly99,color=line,zorder=5)
       
    #21 yard line
        ly98 = [0,88] 
        lx98 = [124,124]
        plt.plot(lx98,ly98,color=line,zorder=5)
        
        ly93 = [0,88] 
        lx93 = [21,21]
        plt.plot(lx93,ly93,color=line,zorder=5)
     
    #45 metre line
        ly96 = [0,88] 
        lx96 = [100,100]
        plt.plot(lx96,ly96,color=line,zorder=5)
        
        ly91 = [0,88] 
        lx91 = [45,45]
        plt.plot(lx91,ly91,color=line,zorder=5)
        
     #65 yard line
        ly95 = [0,88] 
        lx95 = [80,80]
        plt.plot(lx95,ly95,color=line,zorder=5)
        
        ly92 = [0,88] 
        lx92 = [65,65]
        plt.plot(lx92,ly92,color=line,zorder=5)
        
        ## Rectangles in boxes
        rec1 = plt.Rectangle((124,30), 124,30,ls='-',color=pitch, zorder=1,alpha=1)
        rec2 = plt.Rectangle((0, 30), 21,30,ls='-',color=pitch, zorder=1,alpha=1)

        ## Pitch rectangle
        rec3 = plt.Rectangle((-1, -1), 147,90,ls='-',color=pitch, zorder=1,alpha=1)

        ax.add_artist(rec3)
        ax.add_artist(circle1)
        ax.add_artist(circle2)
        ax.add_artist(rec1)
        ax.add_artist(rec2)
       # ax.add_artist(circle3)
        

  #  def onclick(event):
    def onpass(event):
    #def onkey(event):
        global current_coords
        if event.inaxes:
        #if event.button == p:
       # if keyboard.is_pressed('p'):
            current_coords.append((event.xdata, event.ydata))
            if len(current_coords) == 2:
                ax.plot([current_coords[0][0], current_coords[1][0]],
                        [current_coords[0][1], current_coords[1][1]], 'yo-')
                #action = simpledialog.askstring(title="Action_Code", prompt="Action:") 
             #   player = simpledialog.askstring(title="Player", prompt="Player Number:")
              #  receiver = simpledialog.askstring(title="Receiver", prompt="Receiver Number:")               
                coords.append([current_coords[0], current_coords[1], "successful pass"])
                current_coords[:] = []

        #Refresh the plot
            fig.canvas.draw()
            
    def onbadpass(event):
        if event.key == 'u':
            if event.xdata is not None and event.ydata is not None:
                current_coords.append((event.xdata, event.ydata))
                if len(current_coords) == 2:
                    ax.plot([current_coords[0][0], current_coords[1][0]],
                            [current_coords[0][1], current_coords[1][1]], 'ko-')
                  #  action = simpledialog.askstring(title="Action_Code", prompt="Action:") 
                 #   player = simpledialog.askstring(title="Player", prompt="Player Number:")
                  #  receiver = simpledialog.askstring(title="Receiver", prompt="Receiver Number:")
                   # outcome = simpledialog.askstring(title="Outcome", prompt="Outcome:")
                    coords.append([current_coords[0], current_coords[1], "unsuccessful pass"])
                    current_coords[:] = []
                fig.canvas.draw()
                
    def onkickout(event):
        if event.key == 'k':
            if event.xdata is not None and event.ydata is not None:
                current_coords.append((event.xdata, event.ydata))
                if len(current_coords) == 2:
                    ax.plot([current_coords[0][0], current_coords[1][0]],
                            [current_coords[0][1], current_coords[1][1]], 'wo-')
                  #  action = simpledialog.askstring(title="Action_Code", prompt="Action:") 
                 #   player = simpledialog.askstring(title="Player", prompt="Player Number:")
                  #  receiver = simpledialog.askstring(title="Receiver", prompt="Receiver Number:")
                   # outcome = simpledialog.askstring(title="Outcome", prompt="Outcome:")
                    coords.append([current_coords[0], current_coords[1], "successful kickout"])
                    current_coords[:] = []
                fig.canvas.draw()
                
    def onbadkickout(event):
        if event.key == 'c':
            if event.xdata is not None and event.ydata is not None:
                current_coords.append((event.xdata, event.ydata))
                if len(current_coords) == 2:
                    ax.plot([current_coords[0][0], current_coords[1][0]],
                            [current_coords[0][1], current_coords[1][1]], 'ro-')
                  #  action = simpledialog.askstring(title="Action_Code", prompt="Action:") 
                 #   player = simpledialog.askstring(title="Player", prompt="Player Number:")
                  #  receiver = simpledialog.askstring(title="Receiver", prompt="Receiver Number:")
                   # outcome = simpledialog.askstring(title="Outcome", prompt="Outcome:")
                    coords.append([current_coords[0], current_coords[1], "unsuccessful kickout"])
                    current_coords[:] = []
                fig.canvas.draw()

   
    def onaction(event):
   # print(event.key)
        if event.key == 'g':
            if event.xdata is not None and event.ydata is not None:
                ax.plot(event.xdata, event.ydata, 'bo-')
                action = simpledialog.askstring(title="Action_Code", prompt="Action:")
                team = simpledialog.askstring(title="Team", prompt="Team:")
                player = simpledialog.askstring(title="Player", prompt="Player Number:")
                position = simpledialog.askstring(title="Position", prompt="Position:")
                pressure = simpledialog.askstring(title="Pressure", prompt="Pressure:")
                foot = simpledialog.askstring(title="Foot", prompt="Foot:")
                minute = simpledialog.askstring(title="Minute", prompt="Minute:")
            #    player = simpledialog.askstring(title="Player", prompt="Player Number:")
                coords.append([event.xdata, event.ydata, action, team, player, position, foot, minute, pressure])
        
            fig.canvas.draw()
            
    def onbadaction(event):
   # print(event.key)
        if event.key == 'b':
            if event.xdata is not None and event.ydata is not None:
                ax.plot(event.xdata, event.ydata, 'ro-')
                action = simpledialog.askstring(title="Action_Code", prompt="Action:")
                team = simpledialog.askstring(title="Team", prompt="Team:")
                player = simpledialog.askstring(title="Player", prompt="Player Number:")
                position = simpledialog.askstring(title="Position", prompt="Position:")
                pressure = simpledialog.askstring(title="Pressure", prompt="Pressure:")
                foot = simpledialog.askstring(title="Foot", prompt="Foot:")
                minute = simpledialog.askstring(title="Minute", prompt="Minute:")
            #    player = simpledialog.askstring(title="Player", prompt="Player Number:")
             #   outcome = simpledialog.askstring(title="Outcome", prompt="Outcome:")
                coords.append([event.xdata, event.ydata, action, team, player, position, foot, minute, pressure])
        
            fig.canvas.draw()

   
    cid1 = fig.canvas.mpl_connect('key_press_event', onaction)
    cid2 = fig.canvas.mpl_connect('button_press_event', onpass)
    cid3 = fig.canvas.mpl_connect('key_press_event', onbadpass)
    cid4 = fig.canvas.mpl_connect('key_press_event', onbadaction)
    cid5 = fig.canvas.mpl_connect('key_press_event', onkickout)
    cid6 = fig.canvas.mpl_connect('key_press_event', onbadkickout)

    
    plt.show()
    coords
```
