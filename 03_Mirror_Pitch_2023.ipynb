# Read Left to Right

This can be the data from the first or second half. Left to Right refers to the direction of the camera angle on the game. 

```python
df1 = df[df['Half'] == '2nd']
```

# Read Right to Left

```python
df2 = df[df['Half'] == '1st']
```

### Mirror depends on half - mirror to higher x value

```python
df2['mx'] = 72.5
```

```python
df2['my'] = 44
```

```python
df2.head()
```

```python
df2
```

```python
pd.set_option('display.max_rows', 500)
```

```python
df2['x'] = 2*df2['mx'] - df2['x']
df2['y'] = 2*df2['my'] - df2['y']
```

```python
df1['mx'] = 72.5
```

```python
df1['my'] = 44
```

```python
df1['x'] = 2*df1['mx'] - df1['x']
df1['y'] = 2*df1['my'] - df1['y']
```

# Merge frames

```python
frames = (df1, df2)
full = pd.concat(frames)
```
